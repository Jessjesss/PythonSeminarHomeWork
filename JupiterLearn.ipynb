```python
from sympy.plotting import plot
from sympy import *
import math

x = Symbol('x')
f = sin(x)**2 - cos(x)**2

# 1. Построить график
plot(f)
```

```python
# 2. Определить корни f = sin(x)**2 - cos(x)**2
f = sin(x)**2 - cos(x)**2
res = solve(f,x)
print(res)
for i in res:
    print(i.evalf())
```

```python
# 3. Определить промежутки, на котором f > 0

pprint(solveset( f > 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 4. Определить промежутки, на котором f < 0

pprint(solveset( f < 0, x, domain=S.Reals), use_unicode=True)
```

```python
# 5. # Вычислить вершину

difff = diff(f,x)
print(difff)
solve(difff, x)
```

```python
# 6. # Найти интервалы, на которых функция возрастает
     # Найти интервалы, на которых функция убывает

intervals = []
for i in range(-100,100):
    intervals.append(math.sin(i)**2 - math.cos(i)**2)
print(intervals)
```
